# Коэффициент корреляции (КК)

КК показывает силу зависимости между признаками. Чем больше абсолютная величина КК, тем сильнее зависимость. Максимальные абсолютные значения КК -1, 1. В этом случае между признаками существует линейная зависимость

Если КК отрицательный, то между признаками есть обратная зависимость

Если КК очень близок, то фактически никакой зависимости нет.
